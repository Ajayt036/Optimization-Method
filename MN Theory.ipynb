## 12. What Are Feasible and Optimal Solution in LP?

- All possible points under the intersection of lines resulted from constraint equations  are possible value for decision Variables to ultimately find the the value of objective function. They are all called feasible solutions.

- Among all the points or value, One specific point will gives you Max or Min for the objective function. It os your best output and is called optimal Solution. 

- Note that According to corner theorem this point is always lies on intersection of the lines



## 13. What are binding and Non binding Constraints, and Slack?

- When solution of objective function is achieved through final values of Decision Variable, The value from equation constaints can also be calculated. These are often called LHS values.

- If these value are equal to R.H.S Constaints ( value from equation follows inequality), then they are called Binding Consstaints, Othervise Non Binding constraints.

- The Difference between the both sides of constarints is called Slack.

## 15. What Does a Secnsitivity Report From solver tells us?

It Tells us three things:

1. Allowable increase / decrease: How much upto how much increase or decreases in the cofficnets of obj. function does not change out optimal points i.e value of Decision variables. Same way, It also tell us upto how much increase or decreases in the R.H.S Constraints does not change out optimal points i.e value of Decision variable

2. Reduced Cost: A reduced cost in the second column indicates, in general, how much the objective coefficient of a decision variable that is currently 0 must change before the variable changes


3. Shadow Price: A shadow price indicates the change in the optimal value of the objective when the right-hand side of some constraint changes by one unit
